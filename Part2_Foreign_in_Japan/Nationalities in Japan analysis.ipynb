## Immigration by Prefecture
#### 2023/24 Data

1. What are the top 10 immigrant nationalities by entry type in Japan in 2023


```python
pd.read_sql('''
                SELECT *
                FROM immi_entrance
                WHERE year = 2023
                ORDER BY total DESC
                LIMIT 10;
                ''', conn)
```

```python
pd.read_sql('''
                SELECT Prefecture as Country, total
                FROM nat_by_pref
                ORDER BY total DESC
                LIMIT 6;
                ''', conn)
```
